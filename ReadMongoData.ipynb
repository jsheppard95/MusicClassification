```python
# Import dependencies
from pymongo import MongoClient
import gridfs
import pandas as pd
from io import StringIO
# Create a client instance
client = MongoClient()
```

```python
# checkout our client instance
db = client["Music_db"]
```

```python
# Create a gridfs instance
fs = gridfs.GridFS(db)
```

```python
# Create GridFS file instance with fs.get_last_version
features_3_sec_raw = fs.get_last_version("features_3_sec.csv")
```

```python
# Read data as a byte string
bytes_string = features_3_sec_raw.read() 
```

```python
type(bytes_string)
```

```python
# Decode our binary string containing feature data
data_string = bytes_string.decode("utf-8")
data_string
```

```python
# Add feature string to Pandas dataframe
data_string_IO = StringIO(data_string)

features_df = pd.read_csv(data_string_IO, sep=",")
features_df
```
